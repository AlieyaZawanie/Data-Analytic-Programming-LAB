(vi) Consider following rules in classify the attendance:<br>
        a.	Rate > 90% : Good <br>
        b.	60% < Rate < 90% : Moderate <br>
        c.	Rate < 60% : Bad <br>
    Based on the grouping rule above, get the status for each of the staff and save the status in a new column ``status`` in the Data Frame.

```python
# WRITE YOUR CODE HERE

status_attend = []
for i in attendance_rate:
    if i > 90:
        status_attend.append("Good")
    elif i > 60:
        status_attend.append("Moderate")
    else:
        status_attend.append("Bad")

emp_attendance = emp_attendance.assign(status = status_attend)
emp_attendance
```

(vii) Construct a function that returns the average attendance of all staff over the 14 days according to the department. 

```python
# WRITE YOUR CODE HERE

def avg_att_dep(array):
    average_attendance = []
    average_finance = []
    average_staffing = []
    j = 0
    for i in array.total:
        if array.Department[j] == 'FINANCE':
            average_finance.append(i)
        else:
            average_staffing.append(i)
        j = j + 1
            
    average_attendance = [sum(average_finance)/len(average_finance), sum(average_staffing)/len(average_staffing)]
    return average_attendance

print(emp_attendance.Department[2] == 'FINANCE')
average = avg_att_dep(emp_attendance)
final = pd.DataFrame(average, index = ['FINANCE', 'STAFFING'], columns = ["Average Attendance"])
final
```
